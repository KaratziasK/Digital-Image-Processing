```python
import numpy as np
import os
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern, hog
from skimage.io import imread
from skimage.color import rgb2gray
from skimage import exposure
from skimage.transform import resize
from sklearn.metrics import euclidean_distances
import pandas as pd
import matplotlib.pyplot as plt

# ================== A (functions) ===========================
def load_and_convert_image(image_path, target_size=(256, 256)):
    image = imread(image_path)
    gray = rgb2gray(image)
    resized_image = resize(gray, target_size)
    gray_uint8 = (resized_image * 255).astype(np.uint8)
    return gray_uint8

def extract_glcm_features(image, distance=1, angles=[0, np.pi/4, np.pi/2, 3*np.pi/4], levels=256, symmetric=True, normed=True):
    glcm = graycomatrix(
        image,
        distances=[distance],
        angles=angles,
        levels=levels,
        symmetric=symmetric,
        normed=normed
    )
    contrast = graycoprops(glcm, 'contrast')[0, 0]
    correlation = graycoprops(glcm, 'correlation')[0, 0]
    energy = graycoprops(glcm, 'energy')[0, 0]
    homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]

    return np.array([contrast, correlation, energy, homogeneity])

def extract_lbp_features(image):
    lbp = local_binary_pattern(image, P=8, R=1, method='uniform')
    n_bins = int(lbp.max() + 1)
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_bins+1), density=True)
    return lbp, lbp_hist

def extract_hog_features(image):
    fd, hog_image = hog(
        image,
        orientations=9,
        pixels_per_cell=(64, 64),
        cells_per_block=(1, 1),
        visualize=True,
        feature_vector=False
    )
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 1), out_range=(0, 1))
    
    return np.array(fd), np.array(hog_image_rescaled)

def process_dataset(folder_path, target_size=(256, 256)):
    folder_path = os.path.expanduser(folder_path)
    data = []
    classes = os.listdir(folder_path)

    for cls in classes:
        class_path = os.path.join(folder_path, cls)
        if not os.path.isdir(class_path):
            continue
        images = os.listdir(class_path)
        
        for image_name in images:
            image_path = os.path.join(class_path, image_name)
            try:
                image = load_and_convert_image(image_path, target_size)

                glcm_features = extract_glcm_features(image)

                _, lbp_hist = extract_lbp_features(image)

                hog_features, _ = extract_hog_features(image)
                hog_mean = np.mean(hog_features)
                hog_std = np.std(hog_features)

                data.append(list(glcm_features.ravel()) + list(lbp_hist) + [hog_mean, hog_std, cls])

            except Exception as e:
                print(f"Σφάλμα στην εικόνα: {image_path}, {e}")
    return np.array(data, dtype=object)

# ================== B (functions) ===========================
def flatten_2d_to_1d(matrix):
    return matrix.ravel()

def process_dataset_with_flattened_features(folder_path, target_size=(256, 256)):
    folder_path = os.path.expanduser(folder_path)
    data = []
    classes = os.listdir(folder_path)

    for cls in classes:
        class_path = os.path.join(folder_path, cls)
        if not os.path.isdir(class_path):
            continue
        images = os.listdir(class_path)
        
        for image_name in images:
            image_path = os.path.join(class_path, image_name)
            try:
                image = load_and_convert_image(image_path, target_size)

                glcm_features = extract_glcm_features(image)
                glcm_features_flattened = glcm_features.ravel() if glcm_features.ndim == 4 else glcm_features.flatten()

                lbp_image, _ = extract_lbp_features(image)
                lbp_hist, _ = np.histogram(lbp_image.ravel(), bins=np.arange(0, 16+1), density=True)
                lbp_flattened = lbp_hist.ravel()

                hog_features, _ = extract_hog_features(image)
                hog_mean = np.mean(hog_features)
                hog_std = np.std(hog_features)
                hog_features_flattened = np.concatenate([hog_features.ravel(), [hog_mean, hog_std]])

                data.append(list(glcm_features_flattened) + list(lbp_flattened) + list(hog_features_flattened) + [cls])

            except Exception as e:
                print(f"Σφάλμα στην εικόνα: {image_path}, {e}")

    return np.array(data, dtype=object)

# ================== Γ (functions) ===========================
def compute_euclidean_distance(image_path_1, image_path_2):
    gray_image_1 = load_and_convert_image(image_path_1)
    gray_image_2 = load_and_convert_image(image_path_2)

    glcm_features_1 = extract_glcm_features(gray_image_1).ravel().reshape(1, -1)
    glcm_features_2 = extract_glcm_features(gray_image_2).ravel().reshape(1, -1)

    lbp_image_1, _ = extract_lbp_features(gray_image_1)
    lbp_image_2, _ = extract_lbp_features(gray_image_2)

    lbp_hist_1, _ = np.histogram(lbp_image_1.ravel(), bins=np.arange(0, 16+1), density=True)
    lbp_hist_2, _ = np.histogram(lbp_image_2.ravel(), bins=np.arange(0, 16+1), density=True)

    lbp_features_1 = lbp_hist_1.reshape(1, -1)
    lbp_features_2 = lbp_hist_2.reshape(1, -1)

    hog_features_1, _ = extract_hog_features(gray_image_1)
    hog_features_2, _ = extract_hog_features(gray_image_2)

    hog_features_1 = hog_features_1.ravel().reshape(1, -1)
    hog_features_2 = hog_features_2.ravel().reshape(1, -1)

    glcm_distance = euclidean_distances(glcm_features_1, glcm_features_2)[0][0]
    lbp_distance  = euclidean_distances(lbp_features_1, lbp_features_2)[0][0]
    hog_distance  = euclidean_distances(hog_features_1, hog_features_2)[0][0]

    return glcm_distance, lbp_distance, hog_distance

# ================== Δ - Ε (functions) ===========================
def classify_image(image_path, representative_images):
    glcm_distances = []
    lbp_distances = []
    hog_distances = []

    for animal_class, representative_image_path in representative_images.items():
        glcm_distance, lbp_distance, hog_distance = compute_euclidean_distance(image_path, representative_image_path)
        
        glcm_distances.append((animal_class, glcm_distance))
        lbp_distances.append((animal_class, lbp_distance))
        hog_distances.append((animal_class, hog_distance))
    
    glcm_class = min(glcm_distances, key=lambda x: x[1])[0]
    lbp_class = min(lbp_distances, key=lambda x: x[1])[0]
    hog_class = min(hog_distances, key=lambda x: x[1])[0]

    classes = [glcm_class, lbp_class, hog_class]
    final_class = max(set(classes), key=classes.count)

    return final_class, glcm_class, lbp_class, hog_class, glcm_distances, lbp_distances, hog_distances

# ================== A ===========================
check_data = process_dataset(os.path.expanduser('~/code/semester_6/photo/projects/project-2/instructions/archive/chcek/chcek'))
test_data  = process_dataset(os.path.expanduser('~/code/semester_6/photo/projects/project-2/instructions/archive/test/test'))

print("A.Ολοκληρώθηκε η εξαγωγή χαρακτηριστικών!")

# ================== B ===========================
check_data_flattened = process_dataset_with_flattened_features(os.path.expanduser('~/code/semester_6/photo/projects/project-2/instructions/archive/chcek/chcek'))
test_data_flattened  = process_dataset_with_flattened_features(os.path.expanduser('~/code/semester_6/photo/projects/project-2/instructions/archive/test/test'))

print("B.Ολοκληρώθηκε η εξαγωγή χαρακτηριστικών με μετατροπή σε 1Δ διάνυσμα!")

#==================  Γ ==========================
representative_images = {
    'cow': '../instructions/archive/chcek/chcek/cow/cow_1001.jpg',
    'horse': '../instructions/archive/chcek/chcek/horse/horse_1001.jpeg',
    'nilgai': '../instructions/archive/chcek/chcek/Nilgai/nilgai_1001.jpeg',
    'buffalo': '../instructions/archive/chcek/chcek/water buffelo/buffelo_1001.jpeg'
}
test_dirs = [
    '../instructions/archive/test/test/cow',
    '../instructions/archive/test/test/horse',
    '../instructions/archive/test/test/Nilgai',
    '../instructions/archive/test/test/water buffelo'
]

results = []

for animal_class, image_path_1 in representative_images.items():
    print(f"Γ. Comparing images with image: {os.path.basename(image_path_1)}")
    for td in test_dirs:        
        for image_path_2 in os.listdir(td):
            image_path_2_full = os.path.join(td, image_path_2)

            glcm_distance, lbp_distance, hog_distance = compute_euclidean_distance(image_path_1, image_path_2_full)

            results.append({
                'Animal Class': animal_class,
                'Representative Image': os.path.basename(image_path_1),
                'Compared Image': os.path.basename(image_path_2_full),
                'GLCM Distance': glcm_distance,
                'LBP Distance': lbp_distance,
                'HoG Distance': hog_distance
            })

print("Plotting...")
df = pd.DataFrame(results)
fig, axes = plt.subplots(4, 3, figsize=(15, 12))

animal_classes = ['cow', 'buffalo', 'nilgai', 'horse']
for i, animal_class in enumerate(animal_classes):
    animal_df = df[df['Animal Class'] == animal_class]
    
    axes[i, 0].bar(animal_df['Compared Image'], animal_df['HoG Distance'], color='red', alpha=0.7)
    axes[i, 0].set_title(f"{animal_class.capitalize()} - HoG")
    axes[i, 0].tick_params(axis='x', rotation=90)
    axes[i, 0].set_ylabel('HoG Distance')

    axes[i, 1].bar(animal_df['Compared Image'], animal_df['GLCM Distance'], color='blue', alpha=0.7)
    axes[i, 1].set_title(f"{animal_class.capitalize()} - GLCM")
    axes[i, 1].tick_params(axis='x', rotation=90)
    axes[i, 1].set_ylabel('GLCM Distance')

    axes[i, 2].bar(animal_df['Compared Image'], animal_df['LBP Distance'], color='green', alpha=0.7)
    axes[i, 2].set_title(f"{animal_class.capitalize()} - LBP")
    axes[i, 2].tick_params(axis='x', rotation=90)
    axes[i, 2].set_ylabel('LBP Distance')

plt.tight_layout()
plt.show()

#================= Δ - Ε ===========================
test_directory = os.path.expanduser('~/code/semester_6/photo/projects/project-2/instructions/archive/test/test')

glcm_correct = 0
lbp_correct = 0
hog_correct = 0
final_correct = 0
total_images = 0

image_names = []
predictions = []
glcm_preds = []
lbp_preds = []
hog_preds = []
actual_classes = []

print("Δ και Ε")
print("Table...")

for animal_class in os.listdir(test_directory):
    animal_class_path = os.path.join(test_directory, animal_class)
    
    if os.path.isdir(animal_class_path):
        for image_file in os.listdir(animal_class_path):
            image_path = os.path.join(animal_class_path, image_file)
            
            if image_file.lower().endswith(('jpg')):
                final_class, glcm_class, lbp_class, hog_class, _, _, _ = classify_image(image_path, representative_images)

                image_names.append(image_file)
                glcm_preds.append(glcm_class)
                lbp_preds.append(lbp_class)
                hog_preds.append(hog_class)
                predictions.append(final_class)
                actual_classes.append(animal_class)

glcm_accuracy = (sum([1 for i in range(len(actual_classes)) if glcm_preds[i].lower() == actual_classes[i].lower()]) / len(actual_classes)) * 100
lbp_accuracy = (sum([1 for i in range(len(actual_classes)) if lbp_preds[i].lower() == actual_classes[i].lower()]) / len(actual_classes)) * 100
hog_accuracy = (sum([1 for i in range(len(actual_classes)) if hog_preds[i].lower() == actual_classes[i].lower()]) / len(actual_classes)) * 100
final_accuracy = (sum([1 for i in range(len(actual_classes)) if predictions[i].lower() == actual_classes[i].lower()]) / len(actual_classes)) * 100

df_results = pd.DataFrame({
    'Image': image_names,
    'Predicted': predictions,
    'GLCM Prediction': glcm_preds,
    'LBP Prediction': lbp_preds,
    'Hog Prediction': hog_preds,
    'Actual Class': actual_classes
})

accuracy_row = pd.DataFrame({
    'Image': ['Accuracy'],
    'Predicted': [f'{final_accuracy:.2f}%'],
    'GLCM Prediction': [f'{glcm_accuracy:.2f}%'],
    'LBP Prediction': [f'{lbp_accuracy:.2f}%'],
    'Hog Prediction': [f'{hog_accuracy:.2f}%'],
    'Actual Class': ['-']
})
df_results = pd.concat([df_results, accuracy_row], ignore_index=True)

display(df_results)

#====================== ΣΤ ===========================
results = []

representative_images = {
    'cow': '../instructions/archive/test/test/cow/cow_0004.jpg',
    'buffelo': '../instructions/archive/test/test/water buffelo/buffelo_0005.jpg',
    'nilgai': '../instructions/archive/test/test/Nilgai/nilgai_0005.jpg',
    'horse': '../instructions/archive/test/test/horse/horse_0004.jpg'
    
}

print("ΣΤ")

for animal_class, image_path_1 in representative_images.items():
    print(f"Comparing {animal_class} (Representative Image: {image_path_1})")

    for compare_class, representative_image in representative_images.items():
        if animal_class != compare_class:
            glcm_distance, lbp_distance, hog_distance = compute_euclidean_distance(image_path_1, representative_image)

            glcm_distance = round(glcm_distance, 3)
            lbp_distance = round(lbp_distance, 3)
            hog_distance = round(hog_distance, 3)

            results.append({
                'Image': os.path.basename(image_path_1),
                'Class': animal_class,
                'Compared Class': compare_class,
                'GLCM Distance': glcm_distance,
                'LBP Distance': lbp_distance,
                'HoG Distance': hog_distance
            })

df_results = pd.DataFrame(results)

display(df_results)
```
